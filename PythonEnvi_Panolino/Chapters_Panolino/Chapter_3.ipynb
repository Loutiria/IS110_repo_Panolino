#### Exercise 2

Write your salary computation to give an employee 1.5 times the hourly rate for hours worked above 40 hours. Example:

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E3.2

````python

hours_str = input('Enter Hours: ')
rate_str = input('Enter Rate: ')

try:
    hours = float(hours_str)
    rate = float(rate_str)
except ValueError:
    print('Error: Please enter numeric input for hours and rate.')
    exit()

if hours > 40:
    regular_pay = 40 * rate
    overtime_hours = hours - 40
    overtime_pay = overtime_hours * (rate * 1.5)
    total_pay = regular_pay + overtime_pay
else:
    total_pay = hours * rate

print(f'Pay: {total_pay}')
````

## CHALLENGE

Write a program to prompt for a score between 0 and 100 and print a grade using the following table:

    Score           Grade
    [ 98, 100]      1.0
    [ 94,  97)	  	1.25
    [ 90,  93)	    1.5
    [ 86,  89)	    1.75
    [ 82,  85)	    2.0
    [ 78,  81)      2.25
    [ 75,  77)      2.5
    [ 0,   74)	  	5.0

Adapted from [PEW] E3.3
<br>


````python
while True:
    score = input('Enter a score between 0 and 100: ')
    try:
        score_num = float(score)
        if 0 <= score_num <= 100:
            break
        else:
            print('Error: 0 to 100 only.')
    except ValueError:
        print('Error: Invalid input.')

if 98 <= score_num <= 100:
    print('1.0')
elif 94 <= score_num < 98:
    print('1.25')
elif 90 <= score_num < 94:
    print('1.5')
elif 86 <= score_num < 90:
    print('1.75')
elif 82 <= score_num < 86:
    print('2.0')
elif 78 <= score_num < 82:
    print('2.25')
elif 75 <= score_num < 78:
    print('2.5')
else:
    print('5.0')
````
