#### Excerise 1

Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

    Enter Hours: 45
    Enter Rate: 10
    Pay: 475.0
    
[PEW] E4.6

```python
def compute_pay(hours, rate):
    if hours > 40:
        regular_pay = 40 * rate
        overtime_hours = hours - 40
        overtime_pay = overtime_hours * (rate * 1.5)
        total_pay = regular_pay + overtime_pay
    else:
        total_pay = hours * rate
    return total_pay

try:
    hours_worked = float(input("Enter Hours: "))
    hourly_rate = float(input("Enter Rate: "))

    pay = compute_pay(hours_worked, hourly_rate)

    print(f"Pay: {pay}")
except ValueError:
    print("Error: Please enter valid numeric values for hours and rate.")
```

## CHALLENGE

Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

    Score           Grade
    [ 97, 100]      1.0
    [ 93,  97)		1.25
    [ 89,  93)		1.5
    [ 85,  89)		1.75
    [ 81,  85)		2.0
    [ 77,  81)		2.25
    [ 74,  77)		2.5
    [ 0,   74)		3.0

    Run the program repeatedly to test the various different values for input.
    
Adapted from [PEW] E4.6
<br>

```python
def computegrade(score):
    try:
        score_num = float(score)
    except ValueError:
        return "Error: Invalid input."

    if not (0 <= score_num <= 100):
        return "Error: Score out of range. Please enter a score between 0 and 100."

    if 97 <= score_num <= 100:
        return "1.0"
    elif 93 <= score_num < 97:
        return "1.25"
    elif 89 <= score_num < 93:
        return "1.5"
    elif 85 <= score_num < 89:
        return "1.75"
    elif 81 <= score_num < 85:
        return "2.0"
    elif 77 <= score_num < 81:
        return "2.25"
    elif 74 <= score_num < 77:
        return "2.5"
    else:
        return "3.0"

while True:
    user_input = input('Enter a score (or type "done" to quit): ')
    if user_input.lower() == 'done':
        break

    grade = computegrade(user_input)
    print(f"Score: {user_input}, Grade: {grade}")
```
