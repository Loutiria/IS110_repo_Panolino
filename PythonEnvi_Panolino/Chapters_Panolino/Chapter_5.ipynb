#### Exercise 1

Investigate on the difference between `is` and `==` given the following variables.
```python
a = 2
b = 2.0
c = a
```

````python
a = 2
b = 2.0
c = a

print(f"Value of a: {a}, Type of a: {type(a)}, ID of a: {id(a)}")
print(f"Value of b: {b}, Type of b: {type(b)}, ID of b: {id(b)}")
print(f"Value of c: {c}, Type of c: {type(c)}, ID of c: {id(c)}")

print('\n--- Comparisons using == (checks for value equality) ---')
print(f"a == b: {a == b}")
print(f"a == c: {a == c}")
print(f"b == c: {b == c}")

print('\n--- Comparisons using is (checks for identity/same object in memory) ---')
print(f"a is b: {a is b}")
print(f"a is c: {a is c}")
print(f"b is c: {b is c}")
````

#### Challenge 1


Write a program which repeatedly reads numbers until the user enters "done". Once "done" is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number.

    Enter a number: 4
    Enter a number: 5
    Enter a number: bad data
    Invalid input
    Enter a number: 7
    Enter a number: done
    16 3 5.333333333333333
    
[PEW] E5.1

````python
total = 0
count = 0

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == 'done':
        break

    try:
        num = float(user_input)
        total += num
        count += 1
    except ValueError:
        print("Invalid input")
        continue

if count > 0:
    average = total / count
    print(f"\nTotal Sum:{total}\nTotal Count:{count}\nAverage:{average}")
else:
    print("No numbers were entered.")
````

#### Challenge 2

Write another program that prompts for a list of numbers as above and at the end prints out both the maximum and minimum of the numbers instead of the average.

[PEW] E5.2

````python
numbers = []

while True:
    user_input = input("Enter a number: ")

    if user_input.lower() == 'done':
        break

    try:
        num = float(user_input)
        numbers.append(num)
    except ValueError:
        print("Invalid input")
        continue

if len(numbers) > 0:
    max_num = max(numbers)
    min_num = min(numbers)
    print(f"\nMaximum: {max_num}")
    print(f"Minimum: {min_num}")
else:
    print("No numbers were entered.")
````
