## **Exercise 3**

```python
width = 17
height = 12.0

value1 = width // 2
type1 = type(value1)
print(f"\n width // 2: Value = {value1}, \n Type = {type1}")

value2 = width / 2.0
type2 = type(value2)
print(f"\n width / 2.0: Value = {value2}, \n Type = {type2}")

value3 = height / 3
type3 = type(value3)
print(f"\n height / 3: Value = {value3}, \n Type = {type3}")

value4 = 1 + 2 * 5
type4 = type(value4)
print(f"\n 1 + 2 * 5: Value = {value4}, \n Type = {type4}")
```
